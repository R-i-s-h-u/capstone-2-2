```python
import pandas as pd #importing the pandas library
import folium
df=pd.read_csv("D:\downloads\Geospatial_Coordinates.csv")
data=pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")  #reading the data from the website
table=data[0]   #assigning dataframe 
table[table.Borough!='Not assigned'] #removing the not assigned rows from the data
df2=pd.merge(table,df,on="Postal code")#merging the latitude and longitude file with the scrapped data
df2
latitude=43.651070
longitude=-79.347015
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=11)
# add markers to map
for lat, lng, label in zip(df2['Latitude'], df2['Longitude'], df2['Neighborhood']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
       radius=5,
      popup=label,
     color='blue',
    fill=True,
   fill_color='#3186cc',
   fill_opacity=0.7,
   parse_html=False).add_to(map_toronto)  
    
map_toronto
```
